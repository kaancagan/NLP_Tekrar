```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/Users/ayhancagan/Desktop/NPL_ders_tekrar/gender-classifier-DFE-791531.csv", encoding="latin1")
```

```python
data.head(100)
```

```python
#concat ile ihtiyacımız olan verileri cekiyoruz geri kalanını atıyoruz
data = pd.concat([data.gender,data.description],axis=1)
```

```python
data.head()
```

```python
#dropna ile eksik verileri atıyoruz
data.dropna(inplace=True)
print(data.head())
data.reset_index(drop=True, inplace=True)
#reset index yapıyoruz boş olan verilerin yerleri kayıp veri bozulmasın diye
```

```python
data.gender = [1 if i == "female" else 0 for i in data.gender]
print(data.head())
```

```python
#virgüllerden boşluklardan gereksiz şeylerden kurtulmak için kullanıyoruz
import re
metin = re.sub("[^a-zA-Z]", " ", data["description"][9])
print(metin)
harfler = metin.lower()
bol = harfler.split()

```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#stop words yani gereksiz kelimelerden kurtuluyoruz.
import nltk
stop = nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
```

```python
metin = [ps.stem(i)for i in bol if not i in set(stopwords.words("english"))]
metinson = " ".join(metin)# işlemek için metine ceviriyoruz.
```

```python
liste = []
for j in range (1000):
    metin = re.sub("[^a-zA-Z]", " " ,data.description[j])
    metin = metin.lower()
    metin = metin.split()
    metin = [ps.stem(i)for i in metin if not i in set(stopwords.words("english"))]
    metinson = " ".join(metin)
    liste.append(metinson)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vc = CountVectorizer(max_features=10000)
x = vc.fit_transform(liste).toarray()
y = data.iloc[:1000,0].values
```

```python
from sklearn.model_selection import train_test_split

# Verileri bölerken dört ayrı değişkene atanır.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=33)

```

```python
from sklearn.naive_bayes import GaussianNB
gn = GaussianNB()
gn.fit(x_train,y_train)
y_head = gn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_head)
print(cm)
```
